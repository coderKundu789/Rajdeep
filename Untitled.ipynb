```python
import pandas as pd

# Load the Excel file
file_path = 'path_to_your_file.xlsx'
xls = pd.ExcelFile(file_path)

# Extract the header row and relevant data
header_row = 1
df_cleaned = pd.read_excel(file_path, sheet_name='Performance', header=header_row)

# Drop any completely empty columns
df_cleaned.dropna(how='all', axis=1, inplace=True)

# Display the cleaned dataframe
print(df_cleaned.head())

# Assuming the cleaned data is now in df_cleaned, generate SQL create and insert statements
table_name = 'Performance'

# Generate CREATE TABLE statement
columns = df_cleaned.columns
column_definitions = ",\n".join([f"{col.replace(' ', '_')} TEXT" for col in columns])
create_table_sql = f"CREATE TABLE {table_name} (\n{column_definitions}\n);"
print(create_table_sql)

# Generate INSERT INTO statements
insert_sql = []
for index, row in df_cleaned.iterrows():
    values = "', '".join(row.astype(str).replace("'", "''"))
    insert_sql.append(f"INSERT INTO {table_name} VALUES ('{values}');")

# Print INSERT statements
for query in insert_sql:
    print(query)
```
